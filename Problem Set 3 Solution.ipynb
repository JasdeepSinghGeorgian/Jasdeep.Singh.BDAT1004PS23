### Question 1 : 


```python
import pandas as pd

url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, delimiter='|')

print("\n#Step 4 -> Discover what is the mean age per occupation")
print(users.groupby('occupation').age.mean())

print("\n#Step 5 -> Discover the Male ratio per occupation and sort it from the most to the least")
# source: https://stackoverflow.com/questions/30718231/aggregating-in-pandas-groupby-using-lambda-functions
print(users.groupby('occupation')['gender'].agg(lambda x: (x == 'M').sum() / len(x)).sort_values(ascending=False))

print("\n#Step 6a -> Sorting with Minimum Age")
print(users.groupby('occupation').age.min())

print("\n#Step 6b -> Sorting with Maximum Age")
print(users.groupby('occupation').age.max())

print("\n#Step 7 -> For each combination of occupation and sex, calculate the mean age")
print(users.groupby(['occupation', 'gender']).age.mean())


print("\n#Step 8 -> For each occupation present the percentage of women and men")
gender_counts = users.groupby(['occupation', 'gender']).size().unstack()
total_counts = gender_counts.sum(axis=1)
gender_percentages = gender_counts.divide(total_counts, axis=0) * 100
print(gender_percentages)
```

```python
import pandas as pd

url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv'
euro12 = pd.read_csv(url, delimiter=',')

# Step 4: Select only the Goal column
print("\n#Step 4 -> Select only the Goal column")
print(euro12['Goals'])

# Step 5: How many team participated in the Euro2012?
print("\n#Step 5 -> How many team participated in the Euro2012?")
print(euro12['Team'].count())

# Step 6: What is the number of columns in the dataset?
print("\n#Step 6 -> What is the number of columns in the dataset?")
print(len(euro12.columns))

# Step 7: View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline
print("\n#Step 7 -> View only the columns Team, Yellow Cards and Red Cards and assign them to a dataframe called discipline")
dF = euro12[['Team', 'Yellow Cards', 'Red Cards']]
print(dF)

# Step 8: Sort the teams by Red Cards, then to Yellow Cards
print("\n#Step 8 -> Sort the teams by Red Cards, then to Yellow Cards")
print(dF.sort_values(['Red Cards', 'Yellow Cards'], ascending=[False, False]))

# Step 9: Calculate the mean Yellow Cards given per Team
print("\n#Step 9 -> Calculate the mean Yellow Cards given per Team")
print(dF.groupby('Team')['Yellow Cards'].mean())

# Step 10: Teams that scored more than 6 goals
print("\n#Step 10 -> Teams that scored more than 6 goals")
print(euro12[euro12['Goals'] > 6][['Team', 'Goals']].sort_values('Goals'))

#Step 11: Teams that start with G
print("\n#Step 11 -> Teams that start with G")
print(euro12[euro12['Team'].str.startswith('G')]['Team'])

#Step 12: Select the first 7 columns
print("\n#Step 12 -> Select the first 7 columns")
#https://sparkbyexamples.com/pandas/get-first-n-rows-of-pandas/
print(euro12.iloc[:, :7])

#Step 13: Select all columns except the last 3
print("\n#Step 13 -> Select all columns except the last 3")
#https://sparkbyexamples.com/pandas/get-first-n-rows-of-pandas/
print(euro12.iloc[:, :-3])

#Step 14: Present only the Shooting Accuracy from England, Italy and Russia
print("\n#Step 14 -> Present only the Shooting Accuracy from England, Italy and Russia")
print(euro12.loc[euro12['Team'].isin(['England', 'Italy', 'Russia']), ['Team', 'Shooting Accuracy']])
```

### Question 6:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('us-marriages-divorces-1867-2014.csv')
df

df['Marriage Rate'] = (df['Marriages']/df['Population']) * 1000
df['Divorce Rate'] = (df['Divorces']/df['Population']) * 1000

plt.plot(df['Year'], df['Marriage Rate'], label='Marriage Rate')
plt.plot(df['Year'], df['Divorce Rate'], label='Divorce Rate')
plt.xlabel('Year')
plt.ylabel('Per Capita Rate')
plt.title('Marriage and Divorce Rates in the U.S. (1867-2014)')
plt.legend()
plt.show()
```

### Question 7:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('us-marriages-divorces-1867-2014.csv')

df1900 = df[df.Year == 1900]
df1950 = df[df.Year == 1950]
df2000 = df[df.Year == 2000]

combine = [df1900,df1950,df2000]

three_year = pd.concat(combine)
three_year.plot(x = 'Year', y = ['Marriages_per_1000','Divorces_per_1000'], kind = 'bar')
plt.xlabel('Year')
plt.ylabel('Rate per 1000 population')
plt.title('Marriages and Divorces in the U.S. (1900, 1950, and 2000)')
plt.legend(['Marriages', 'Divorces'])
plt.show()
```

### Question 9:

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('roman-emperor-reigns.csv')

# Count the number of emperors and the number of emperors who were assassinated
total_emperors = len(df)
assassinated_emperors = len(df[df['Cause_of_Death'] == 'Assassinated'])

# Compute the fraction of emperors who were assassinated
fraction_assassinated = assassinated_emperors / total_emperors

# Compute the fraction of emperors who were not assassinated
fraction_not_assassinated = 1 - fraction_assassinated

# Create a pie chart
labels = ['Assassinated', 'Not Assassinated']
sizes = [fraction_assassinated, fraction_not_assassinated]
colors = ['red', 'cyan']

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
ax.set_title('Fraction of Roman Emperors Assassinated')
plt.show()
```
